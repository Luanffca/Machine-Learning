```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/clientes.csv')
print(df.info())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
codificador = LabelEncoder()
df['profissao'] = codificador.fit_transform(df['profissao'])
df['mix_credito'] = codificador.fit_transform(df['mix_credito'])
df['comportamento_pagamento'] = codificador.fit_transform(df['comportamento_pagamento'])
df
```

```python
print(df.info())
```

```python
y = df['score_credito']
x = df.drop(['score_credito', 'id_cliente'], axis=1)

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
# Criando o modelo de IA -> Score de credito: Ruim, Medio, Bom
# Modelo de Arvore de Decisao

from sklearn.ensemble import RandomForestClassifier
modelo_arvoredecisao = RandomForestClassifier()
modelo_arvoredecisao.fit(x_treino, y_treino)
```

```python
# KNN - nearest neighbords
from sklearn.neighbors import KNeighborsClassifier
modelo_KNN = KNeighborsClassifier()
modelo_KNN.fit(x_treino, y_treino)
```

```python
# Acuracia dos modelos

from sklearn.metrics import accuracy_score

previsao_arvoredecissao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_KNN.predict(x_teste)
```

```python
print("{:.2f}% modelo -> árvore de decisão".format(accuracy_score(y_teste, previsao_arvoredecissao)*100))
print("{:.2f}% modelo -> KNN".format(accuracy_score(y_teste, previsao_knn)*100))
```
