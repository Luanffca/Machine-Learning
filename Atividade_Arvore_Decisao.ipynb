```python
# Importando as Bibliotecas 
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
from sklearn import metrics
```

```python
# Retornando os Dados
dataset_iris = load_iris()
X = dataset_iris.data
y = dataset_iris.target
```

```python
arvore_decisao =  tree.DecisionTreeClassifier(criterion="entropy", max_depth=4)
arvore_decisao.fit(X, y)
```

```python
metrics.accuracy_score(y,arvore_decisao.predict(X))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(30,30))
tree.plot_tree(arvore_decisao, feature_names=dataset_iris.feature_names, class_names=dataset_iris.target_names, filled=True)
plt.show()
```

```python
novo_item = [5,3,1,0.2]
previsao = arvore_decisao.predict([novo_item])
dataset_iris.target_names[previsao]
```

```python
# Transformando em um DataFrame
df = load_iris()
iris = pd.DataFrame(df.data)
iris.columns = df.feature_names
iris['target'] = df.target
```

```python
# separando X e Y da base de dados
X = iris.drop('target', axis=1)
y = iris.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=33, random_state=1)
```

```python
# Criando o Classificador e fazendo o fit
clf2 = tree.DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
```

```python
# Verificando Score
clf2.score(X_train, y_train)
```
